```python
# An iterator is an object that allows you to traverse through a collection of elements one by one without exposing the underlying structure,

li = iter([1,2,3,4,5])
print(type(li))

while True:
    try:
        n = next(li)
        print(n)
    except StopIteration:
        print("your iteration was done ")
        break
```

```python
# generator
def generator():
    yield 'a'
    yield 'b'
    yield 'c'

b = generator()

print(next(b))
print(next(b))
print(next(b))
```

```python
# Given a string, use iter() to create an iterator and print
# its characters one by one using next().

str = iter("ayush")
while True:
    try :
        n = next (str)
        print(n)
    except StopIteration:
        pass
        break
```

```python
# 3. Tuple Iteration: Write a program to iterate through the elements of a tuple
# using iter() and next() without using a for loop.

tup = iter((10, 20, 30, 40, 50))


while True:
    try:
        value = next(tup)
        print(value)
    except StopIteration:
        pass
        break
```

```python
# 4. Dictionary Iteration: Demonstrate how to iterate through a dictionary's
# keys and values simultaneously using iter() and next() on the results of
# the dictionary's .items() method.

dic = iter ({"name":"ayush" , "age":19})

while True :
    try :
        n = next(dic)
        print(n)
    except StopIteration:
        pass
        break
```

```python
# 5. Filtered Iteration: Use an iterator created from a given list to print only the
# even numbers. You must use next() to move through the iterator.
li = iter([1,2,3,4,5,6,7,8,9])

while True:
    try:
        n = next(li)
        if n%2 == 0 :
            print(n)
    except StopIteration:
        pass
        break
            
```

```python
# 6. Partial Iteration: Convert a list into an iterator and print the first three
# elements explicitly using next() three times.
lst = iter([10, 20, 30, 40, 50])


print(next(lst))  
print(next(lst))  
print(next(lst))  
```

```python
# 7. Separate States: Create two distinct iterators for the same initial list. Use
# next() on both iterators in an alternating fashion (e.g., next(iter1),
# next(iter2), next(iter1)) to demonstrate that each iterator maintains a
# separate position (state).

lst1 = iter([10, 20, 30, 40, 50])
lst2 = iter([10, 20, 30, 40, 50])

print("lst1:", next(lst1))   # 10
print("lst2:", next(lst2))   # 10
print("lst1:", next(lst1))   # 20
print("lst2:", next(lst2))   # 20
print("lst1:", next(lst1))   # 30
```
