```python
# lamda function
# high level 


#    MAP 

def square(num):
    return num * num

numbers = [2, 4, 6, 8]
result = map(square, numbers)
print(list(result))  # Output: [4, 16, 36, 64]
```

```python
#                                         FILTER 
#    FILTER FUNCTION SHOWS DIRECT VALUE WHICH IS SITUABLE OF CONDITION 

newlist=[1,2,3,4,5,6,7,8]
out_map2 = list( filter( lambda data: data % 2 == 0  , newlist ) )
print(out_map2)
```

```python
# Write some numbers in a file
f = open("numbers.txt", "w")
f.write("5\n10\n13\n20\n29\n30\n")
f.close()

# Read numbers from file
f = open("numbers.txt", "r")
nums = f.readlines()
f.close()

# Open another file to write results
f = open("prime_results.txt", "w")

for n in nums:
    n = int(n.strip())  # Remove \n and convert to int
    if n < 2:
        f.write(f"{n} is Not Prime\n")
    else:
        for i in range(2, n):
            if n % i == 0:
                f.write(f"{n} is Not Prime\n")
                break
        else:
            f.write(f"{n} is Prime\n")

f.close()

print("✅ Results saved in prime_results.txt")
```
